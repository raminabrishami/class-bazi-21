```python
import random
from IPython.display import clear_output
```

```python
class mablagh_koll():
    
    def __init__(self,mablagh_kol=0):
        self.mablagh_kol=int(mablagh_kol)
        self.mablagh_kol_bazi()

        
    def mablagh_kol_bazi(self):
        global mablagh_kol
        while True:
            try:
                self.mablagh_kol+=int(input('lotfan kol sarmaye khod ra vared konid : '))
                mablagh_kol=self.mablagh_kol
                break
            except:
                clear_output()
                print('lotfan adad vared konid')
                
    def mablaagh(self):
        global mablagh
        while True:
            try:         
                mablagh=int(input('lotfan mablagh shart bazi ra vared konid : '))
                break
            except:
                clear_output()
                print('lotfan adad vared konid')
    
    
    def mablagh_kool(self):
        global halghe_kol
        l=True
        while l==True:
                z=input('mojodi hesab shoma kam ast aya sharje mikonid, Yes or No ? ')
                z=z.lower()
                if z=='yes':
                    self.mablagh_kol_bazi()
                    break
                elif z=='no':
                    print('nemitavanid bazi konid')
                    halghe_kol=False
                    break
                else:
                    print('eshtebah vared kardid') 
                    print('=============================')
                 
```

```python
class mohaseb():
    def __init__(self):
        self.jam_emtiyaz_bazikon()
        self.jam_emtiyaz_pc()
        self.bakhtan_bazikon()
        self.bakhtan_pc()
    
    
    def jam_emtiyaz_bazikon(self):
        global kart_bazikon
        global emtiyazz
        emtiyaz_bazikon=0
        emtiyazz=0
        jaz=-1
    

        for i in range(len(kart_bazikon)):
            emtiyazz+=emtiyaz_kartha[kart_bazikon[i][1]]
            emtiyaz_bazikon+=emtiyaz_kartha[kart_bazikon[i][1]]
        
                
            if emtiyazz>21:
                for j in range(len(kart_bazikon)):
                    if kart_bazikon[j][1]=='1':
                        if jaz!=j:
                            jaz=j
                            if emtiyazz>21:
                                emtiyaz_bazikon-=emtiyaz_kartha[kart_bazikon[j][1]]
                                emtiyazz-=emtiyaz_kartha[kart_bazikon[j][1]]
                                emtiyaz_bazikon+=1
                                emtiyazz+=1
    
            
        return(emtiyaz_bazikon)
    
    
    def jam_emtiyaz_pc(self):
        global kart_pc
        emtiyaz_pc=0
        emtiyazz=0
    
        for i in range(len(kart_pc)):
            emtiyazz+=emtiyaz_kartha[kart_pc[i][1]]
            emtiyaz_pc+=emtiyaz_kartha[kart_pc[i][1]]
        
                
            if emtiyazz>21:
                for j in range(len(kart_pc)):
                    if kart_pc[j][1]=='1':
                        if jaz!=j:
                            jaz=j
                            if emtiyazz>21:
                                emtiyaz_pc-=emtiyaz_kartha[kart_pc[j][1]]
                                emtiyazz-=emtiyaz_kartha[kart_pc[j][1]]
                                emtiyaz_pc+=1
                                emtiyazz+=1
            
        return (emtiyaz_pc)
    
    
    
    def bakhtan_bazikon(self):
        if self.jam_emtiyaz_bazikon()>21:
            return False
        else:
            return True
        
    def bakhtan_pc(self):
        if self.jam_emtiyaz_pc()>21:
            return False
        else:
            return True
    
    
    def moghayese_bordan(self):
        
        global bord
        if self.bakhtan_bazikon()==True and self.bakhtan_pc()==True:
            if self.jam_emtiyaz_bazikon()>self.jam_emtiyaz_pc():
                print('bazikon bord')
                bord='bazikon'
            elif self.jam_emtiyaz_bazikon()<self.jam_emtiyaz_pc():
                print('pc bord')
                bord='pc'
            else:
                print('mosavi')
                bord='mosavi'
        else:
            if self.bakhtan_bazikon()==False:
                print('pc bord')
                bord='pc'
            elif self.bakhtan_pc()==False:
                print('bazikon bord')
                bord='bazikon'
    
    def mohasebat(self):
        global mablagh_kol
        global bord
        global mablagh
        global kart
        l=True
        if bord=='bazikon':
            mablagh_kol=mablagh_kol+mablagh
            return (mablagh_kol)
        elif bord=='pc':
            mablagh_kol=mablagh_kol-mablagh
            return(mablagh_kol)
        elif bord=='mosavi':
            return (mablagh_kol)
        

    def bazi_mojadad(self):
        global halghe_kol
        while True:
            j=input('dobare bazi mikonid, Yes or No ? ')
            j=j.lower()
            if j=='no':
                halghe_kol=False
                break
            elif j=='yes':
                halghe_kol=True
                break
            else :
                print('lotfan Yes ya No entekhab konid')
        
    
```

```python
class dadan_kart():
    def __init__(self):
        self.dadan_kart_pc()
        self.dadan_kart_bazikon()
        c=mohaseb()
        
    def random_chose(self):
        x=random.choice(kart)
        for i in kart:
            if i==x:
                kart.remove(i)
        return x
    
    def dadan_kart_pc(self):
        global kart_pc
        kart_pc=[]
        for i in range(2):
            kart_pc.append(self.random_chose())
        return kart_pc
    
    def dadan_kart_bazikon(self):
        global kart_bazikon
        kart_bazikon=[]
        for i in range (2):
            kart_bazikon.append(self.random_chose())
        return kart_bazikon
    
    def kart_khastan_pc(self):
        c=mohaseb()
        global kart_pc
        while True:
            if c.jam_emtiyaz_pc() < c.jam_emtiyaz_bazikon() :
                kart_pc.append(self.random_chose())
            else:
                break
        if c.bakhtan_pc()==False:
            print('jam emtiyaz pc bishtar az 21 shod ')
            print('=============================')
               

    def kart_khastan_bazikon(self):
        c=mohaseb()
        halghe=True
        global mablagh
        global kart_bazikon
        while halghe==True:
            u=input('aya mikhahid kart bekeshid, Yes or No ? ')
            u=u.lower()
            if u=='yes':
                clear_output()
                kart_bazikon.append(self.random_chose())
                print('kart ha: ',kart_bazikon)
                print('=============================') 
                print('jam emtiyaz: ',c.jam_emtiyaz_bazikon())
                if c.bakhtan_bazikon()==False:
                    clear_output()
                    print('jam emtiyaz shoma bishtar az 21 ahod')
                    print('=============================')
                    c.moghayese_bordan()
                    halghe=False
            elif u=='no':
                clear_output()
                self.kart_khastan_pc()
                c.moghayese_bordan()
                halghe=False
            else:
                print('lotfan Yes ya No entekhab konid')
                print('=============================')
                
        print('=============================')    
        print('kart haye bazikon: ' , kart_bazikon)
        print('=============================')
        print('jam emtiyaz bazikon: ' , c.jam_emtiyaz_bazikon())
        print('=============================')
        print('kart haye pc: ' , kart_pc)
        print('=============================')
        print('jam emtiyaz pc: ' , c.jam_emtiyaz_pc())
        print('=============================')
        print('mablagh shart bandi: ' , mablagh)
        print('=============================')
        print('mablagh kol sarmaye baghi mande: ' , c.mohasebat())
        print('=============================')
    
            
    
```
